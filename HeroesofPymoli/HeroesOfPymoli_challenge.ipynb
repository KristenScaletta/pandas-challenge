## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns

* Create a summary data frame to hold the results

* Display a preview of the summary data frame



```python
popular_df = data_file_df[['Item ID', 'Item Name', 'Price']]
popular_df = popular_df.groupby(['Item ID', 'Item Name'])
popular_df.count().head()
popular_analysis_df = pd.DataFrame({
    "Purchase Count": popular_df.count().head()['Price'],
    "Average Item Price": round(popular_df.mean().head()['Price'], 2),
    "Total Purchase Value": round(popular_df.sum().head()['Price'], 2)
})
popular_analysis_df.sort_values("Purchase Count", ascending=False)
```
