```python
# Dependencies and Setup
import pandas as pd
import os

# Indicates file to load
data_file = os.path.join("Resources", "02-Homework_04-Pandas_Instructions_HeroesOfPymoli_Resources_purchase_data.csv")

# Reads File and stores into Pandas data frame
data_file_df = pd.read_csv(data_file)
data_file_df.head()
```

## Player Count

* Display the total number of players


```python
#Determines how many players there are by calculating the total number of unique values for the screen name field.
unique_players_count = len(data_file_df["SN"].unique())

#Displays these results in a new dataframe.
df_players = pd.DataFrame({
    "Total Number of Players": [unique_players_count]
    })
#Displays the new dataframe
df_players.head()

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Adds variables to store calculations such as number of unique items, average price, etc.
unique_items = len(data_file_df["Item ID"].unique())
average_price = round(data_file_df["Price"].mean(),2) 
number_purchases = len(data_file_df["Purchase ID"])
total_revenue = round(data_file_df["Price"].sum(),2)

#Creates a new summary dataframe to display data calculated above.
summary_data_df = pd.DataFrame({
    "Number of Unique Items":[unique_items],
    "Average Price": [average_price], 
    "Number of Purchase": [number_purchases],
    "Total Revenue": [total_revenue] 
    })

#Displays the summary dataframe.
summary_data_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Gets unique players by pulling data that is constant across each row from the main file and dripping duplicate rows.
unique_players_df = data_file_df[['SN', 'Age', 'Gender']]
unique_players_df = unique_players_df.drop_duplicates()
#Calculates total players to use in percentage calculation.
total_players = unique_players_df['SN'].count()

#Deterimines number of players by gender and the percentage of total players and stores this data in a new dataframe

gender_df = pd.DataFrame({
    "Total Count": unique_players_df['Gender'].value_counts(),
    "Percentage of Players": round(unique_players_df['Gender'].value_counts()/total_players * 100,2)
})

#Displays the new dataframe.
gender_df

    




    

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#WORK ON
#Creates a dataframe that groups the main data file by gender.
grouped_gender_df = data_file_df.groupby(['Gender'])

#Creates a new dataframe that runs calculations on various summary data pieces and stores them by gender.
gender_purchase_df = pd.DataFrame({
    "Purchase Count": grouped_gender_df.count()['Item ID'],
    "Average Purchase Price": round(grouped_gender_df.mean()['Price'], 2),
    "Total Purchase Value": round(grouped_gender_df.sum()['Price'], 2),
    "Avg Total Purchase per Person": round(grouped_gender_df.sum()['Price']/unique_players_df['Gender'].value_counts(),2)
})

#Outputs the sumamry dataframe.
gender_purchase_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Establish bins and groups for different age ranges
bins = [0, 9, 14, 19, 24, 29, 34, 39, 300]
group_by = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorizes players in age bins
data_file_df["Age Summary"] = pd.cut(data_file_df["Age"], bins, labels=group_by, include_lowest=True)
#age_counts = unique_players_df["Age Summary"].value_counts()

#Groups dataframe by Age Summary
grouped_age_df = data_file_df.groupby(['Age Summary'])

#Creates a new dataframe that runs summary calculations and displays them. Note-this calls a dataframe and column created in an above cell so that must be run first.
purchase_analysis_df = pd.DataFrame({
    "Purchase Count": grouped_age_df.count()['Item ID'],
    "Average Purchase Price": round(grouped_age_df.mean()['Price'], 2),
    "Total Purchase Value": round(grouped_age_df.sum()['Price'], 2),
    "Avg Total Purchase per Person": round(grouped_age_df.sum()['Price']/unique_players_df['Age Summary'].value_counts(),2)
})
purchase_analysis_df
#data_file_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Groups the main data file by SN to run calculations by individual.
grouped_spender_df = data_file_df.groupby(['SN'])

#Calculates total purchases, average purchase price and total purchase value and adds them as values to the corresponding keys in a new dataframe.
purchase_analysis_df = pd.DataFrame({
    "Purchase Count": grouped_spender_df.count()['Item ID'],
    "Average Purchase Price": round(grouped_spender_df.mean()['Price'], 2),
    "Total Purchase Value": round(grouped_spender_df.sum()['Price'], 2),
    })
#Sorts the data by Total Purchase Value with the highest value being at the top.
purchase_analysis_df.sort_values("Total Purchase Value", ascending=False)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieves select columns in a new dataframe
popular_df = data_file_df[['Item ID', 'Item Name', 'Price']]

#Groups the new dataframe by Item ID and Item Name
popular_df = popular_df.groupby(['Item ID', 'Item Name'])

#Calculates the purchase count, average item price and total purchase value and outputs this into a summary data frame.
popular_analysis_df = pd.DataFrame({
    "Purchase Count": popular_df.count()['Price'],
    "Average Item Price": round(popular_df.mean()['Price'], 2),
    "Total Purchase Value": round(popular_df.sum()['Price'], 2)
})

#Sorts the summary data frame by purchase count, with the highest being on top.
popular_analysis_df.sort_values("Purchase Count", ascending=False)
```
